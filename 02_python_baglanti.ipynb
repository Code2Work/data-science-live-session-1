# Bolum 2: Python ile Veritabanina Baglanmak

Python'dan PostgreSQL'e nasil baglanilir?

`psycopg2` kutuphanesi ile baglanti kurup SQL sorgulari calistirabiliyoruz.

## Kutuphaneyi yukleyelim

```python
import psycopg2
```

## PostgreSQL'e baglanma
Baglanti bilgilerini kendi bilgisayariniza gore degistirin.

```python
conn = psycopg2.connect(
    host="localhost",
    database="postgres",
    user="postgres",
    password="1234"
)
print("Baglanti basarili!")
```

## Test — kac urunumuz var?

```python
cursor = conn.cursor()
cursor.execute("SELECT COUNT(*) FROM products")
print("Toplam urun sayisi:", cursor.fetchone()[0])
```

## Birden fazla satir cekmek

```python
cursor.execute("""
    SELECT category_name, COUNT(product_id)
    FROM products p
    JOIN categories c ON p.category_id = c.category_id
    GROUP BY category_name
    ORDER BY count DESC
""")

print("Kategorilere gore urun sayisi:")
for row in cursor.fetchall():
    print(f"  {row[0]}: {row[1]} urun")
```

**Not:** `cursor` ile veri cekmek calisir ama sonuc duz tuple'lar olarak gelir.
